# Aggression Baselines

Each aggression score is in the ordered set [-3, -2, -1, 0, 1, 2, 3]. For each revision, we have k scores that we can aggregate to get a label. Depending on how we aggregate, we can get a regression or a multi-class classification problem. We will build baselines for each setting for both the random and the blocked data.

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
from baselines import *
```

```python
data_filename  = '../data/v4_annotated/annotated_onion_layer_5_rows_0_to_5000_raters_20.csv'
d_b = pd.read_csv(data_filename)
d_b = d_b.query('_golden == False')
d_b.index = d_b.rev_id
l_b = d_b['aggression_score']
```

```python
data_filename  = '../data/v4_annotated/annotated_random_data_rows_0_to_5000_raters_20.csv'
d_r = pd.read_csv(data_filename)
d_r = d_r.query('_golden == False')
d_r.index = d_r.rev_id
l_r = d_r['aggression_score']
```

```python
k = 20
```
