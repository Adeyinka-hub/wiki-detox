```python
import pandas as pd
import numpy as np
import random
from diff_utils import print_block_data, print_user_history 
```

```python
# download dat from google drive: wikipdia data/v2
d = pd.read_csv('../data/all_no_admin_blocked_user_long_sample.tsv', sep = '\t')
```

```python
counts = d['user_text'].value_counts()
print ('IQR: ', np.percentile(counts, [25,75 ]))
```

```python
# select random sample of users who have made between 5 and 25 comments
# Note: I fudged the range a bit
IQR_user = counts[counts < 25]
IQR_user = IQR_user[IQR_user > 5]
IQR_user = IQR_user.index.values
random.seed(0)
sample = random.sample(list(IQR_user), 5)
print(sample)
```

```python
# User histories for each selected users
for user in sample:
    print_user_history(d, user)
    print('\n'*2)
```
