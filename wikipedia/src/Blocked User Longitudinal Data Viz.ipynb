```python
import pandas as pd
import numpy as np
import random
```

```python
# download dat from google drive: wikipdia data/v2
d = pd.read_csv('../data/all_no_admin_blocked_user_long_sample.tsv', sep = '\t')
```

```python
counts = d['user_text'].value_counts()
print ('IQR: ', np.percentile(counts, [25,75 ]))
```

```python
# select random sample of users who have made between 5 and 25 comments
# Note: I fudged the range a bit
IQR_user = counts[counts < 25]
IQR_user = IQR_user[IQR_user > 5]
IQR_user = IQR_user.index.values
random.seed(0)
sample = random.sample(list(IQR_user), 5)
print(sample)
```

```python

def print_user_history(d, user):
    
    """
    Print out users comments in order
    """
    
    d_u = d[d['user_text'] == user].sort_values(by ='rev_timestamp')
    
    print ('#' * 80)
    print('History for user: ', user)
    print('\nBlock History')
    
    for c in ['block_reasons', 'block_timestamps', 'block_actions\n']:
        print(c.strip(), ': ', d_u.iloc[0][c].strip())
    print ('#' * 80)
    
    
    for i, r in d_u.iterrows():
        print ('\n')
        print ('User: ', r['user_text'])
        print ('User Talk Page: ', r['page_title'])
        print('Timestamp: ', r['rev_timestamp'])
        print ('\n')
        print (r['clean_diff'])
        print ('_' * 80)
```

```python
# User histories for each selected users
for user in sample:
    print_user_history(d, user)
    print('\n'*2)
```
