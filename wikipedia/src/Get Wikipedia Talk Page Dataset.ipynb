# Just Query a Bunch of Talk page revisions

```python
n = 10

query = """
SELECT
    CAST(old_text AS CHAR(255) CHARSET utf8)  as comment,
    page_id,
    page_namespace,
    page_title,
    rev_comment,
    rev_id,
    rev_minor_edit as minor_edit,
    rev_timestamp as timestamp,
    rev_user as user_id,
    rev_user_text as user_text
FROM
    enwiki.revision r,
    enwiki.text t,
    enwiki.page p 
WHERE
    r.rev_page = p.page_id
    AND r.rev_text_id = t.old_id
    AND rev_timestamp > 20150722165104
LIMIT %d
"""

d = query_analytics_store(query % n, {})
```
